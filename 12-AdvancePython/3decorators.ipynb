```python
###decorators

def my_decorator (func):
    def wrapper ():
        print ("before calling the function")
        func()
        print ("after calling the function")
    return wrapper()
```

```python
def say_hello():
    print("This is an advanced python course covering decorators, generators, and more.")
say_hello()
```

```python
### decorators with arguments

def repeat(n):
    def decorator (func):
        def wrapper (*args, **kwargs):
            for _ in range (n):
                func (*args, **kwargs)
            return wrapper
        return decorator
```

```python
@repeat(3)
def say_hello ():
    print ("Hello")
```

```python
say_hello()
```
